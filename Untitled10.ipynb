```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('img.png')
mask = np.zeros(img.shape[:2],np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (90,100,450,290)
cv.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
plt.imshow(img),plt.show()
```

```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt

```

```python
img = cv2.imread('v2.png')
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
ret ,thresh =cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python

kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
plt.imshow(opening)


sure_bg = cv2.dilate(opening,kernel,iterations=3)
plt.figure()
plt.imshow(sure_bg)



dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
plt.figure()
plt.imshow(sure_fg)
plt.imsave('v6.png',sure_bg)


sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
plt.figure()
plt.imshow(unknown)
```

```python
originalImg = cv2.imread('v2.png',0)
mask = cv2.imread('v6.png',0)
mask[mask < 128] = 0
mask[mask > 128] = 1
outputImg = originalImg * (mask ==1)

outputImg[mask==0] = 255
plt.imshow(outputImg)

cv2.imwrite('new2.png',outputImg)
```

```python
image = cv2.imread("f.jpg") 
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image)
```

```python
weaker = np.array([20,100,100])
stronger = np.array([35,255,255])
```

```python
mask = cv2.inRange(hsv, weaker, stronger) 

cv2.imshow('Image',image)
cv2.imshow('Result',mask) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```
